```python
X = df.drop("Exam_Score", axis=1)
y = df["Exam_Score"]
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

print(num_cols)
print(cat_cols)
```

```python
len(num_cols)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=40)
```

## Preprocessing pipeline

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median"))
])
```

```python
cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])
```

```python
X_train_prepared = preprocessor.fit_transform(X_train)
X_test_prepared = preprocessor.transform(X_test)
```

# Model Building (linear Regression)

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_prepared, y_train)
```

```python
y_pred = model.predict(X_test_prepared)
```

# Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python


results = pd.DataFrame({
    "Actual": y_test,
    "Predicted": y_pred
})

results.head(10)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

dt = DecisionTreeRegressor(random_state=42)

dt.fit(X_train_prepared, y_train)

y_pred_dt = dt.predict(X_test_prepared)

print("Decision Tree R²:", r2_score(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train_prepared, y_train)

y_pred_rf = rf.predict(X_test_prepared)

print("Random Forest R²:", r2_score(y_test, y_pred_rf))
```

```python
comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest"],
    "R2 Score": [
        r2_score(y_test, y_pred),
        r2_score(y_test, y_pred_dt),
        r2_score(y_test, y_pred_rf)
    ]
})

comparison.sort_values(by="R2 Score", ascending=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_prepared, y_train)
y_pred_rf = rf.predict(X_test_prepared)

print("R² Score:", r2_score(y_test, y_pred_rf))
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

scores = cross_val_score(
    LinearRegression(),
    X_train_prepared,
    y_train,
    cv=5,
    scoring="r2"
)

print(scores)
print("Mean R²:", scores.mean())
```

Conclusion

- Three regression models were trained and evaluated.
- Linear Regression achieved the highest R² score of 0.7159.
- Decision Tree performed poorly with an R² score of 0.143.
- Random Forest improved over Decision Tree but achieved only 0.639.
- Therefore, Linear Regression was selected as the final model for predicting student exam scores.
